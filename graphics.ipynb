```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
news        = ['fake 5','fake 4','fake 3','fake 2','fake 1','real 5','real 4','real 3','real 2','real 1','article 5','article 4','article 3','article 2','article 1']
```

### k= 75

```python
resultOne   = [0.001,0.010,0.016,0.078,0.061,0.157,0.272,0.213,0.295,0.442,0.699,0.395,0.291,0.730,0.394]
resultTwo   = [0.001,0.010,0.016,0.079,0.061,0.156,0.269,0.202,0.293,0.439,0.698,0.393,0.286,0.724,0.393]
resultThree = [0.001,0.009,0.016,0.080,0.053,0.156,0.268,0.202,0.291,0.439,0.689,0.390,0.282,0.718,0.388]
```

```python
x=np.arange(15)
plt.figure(figsize=(14,8))
plt.bar(x+0.0,resultOne,width=0.20,color='red',label='5000 documentos')
plt.bar(x+0.25,resultTwo,width=0.20,color='darkblue',label='6000 documentos')
plt.bar(x+0.50,resultThree,width=0.20,color='green',label='7000 documentos')
plt.legend()
plt.title('Predicción con k=75')
plt.xticks(x,news)
plt.xlabel('noticias')
plt.ylabel('prediccion')
plt.savefig('75.png')
plt.show()
```

```python
x=np.arange(15)
plt.figure(figsize=(14,8))
plt.plot(news,resultOne,'-o',color='red',label='5000 documentos')
plt.plot(news,resultTwo,'-o',color='orange',label='5000 documentos')
plt.plot(news,resultThree,'-o',color='yellowgreen',label='5000 documentos')
plt.legend()
plt.title('Predicción con k=75')
plt.xlabel('noticias')
plt.ylabel('prediccion')
plt.savefig('75.png')
plt.show()
```
